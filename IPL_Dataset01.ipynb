```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("G:\matches.csv")
df
```

```python
print(df.isnull().sum()) 
```

```python
df.loc[(df['city'].isna()) & (df['venue'] == 'Sharjah Cricket Stadium'), 'city'] = 'Sharjah'
df.loc[(df['city'].isna()) & (df['venue'] == 'Dubai International Cricket Stadium'), 'city'] = 'Dubai'
df['city'].isnull().sum()
```

```python
df['player_of_match']=df['player_of_match'].fillna('no_player_of_match')
```

```python
df['player_of_match'].isnull().sum()
```

```python
df['winner']=df['winner'].fillna('no_winner')
```

```python
df['winner'].isnull().sum()
```

```python
df['result_margin'].dtype
df['result_margin'] = df['result_margin'].fillna(0).astype(int) 
df['result_margin']
df['result_margin'].isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df_deliveries=pd.read_csv("G:\deliveries.csv")
df_deliveries
```

```python
print(df_deliveries.isnull().sum())
```

```python
df_deliveries['extras_type']=df_deliveries['extras_type'].fillna('no_extras') 
```

```python
df_deliveries['player_dismissed']=df_deliveries['player_dismissed'].fillna('no player dismissed') 
```

```python
df_deliveries['player_dismissed'].isnull().sum()
```

```python
df_deliveries['dismissal_kind']=df_deliveries['dismissal_kind'].fillna('no dismissal')
```

```python
df_deliveries['dismissal_kind'].isnull().sum()
```

```python
df_deliveries['fielder']=df_deliveries['fielder'].fillna('no fielder') 
```

```python
df_deliveries['fielder'].isnull().sum()
```

```python
df_deliveries.isnull().sum()
```

```python
num_rows = df_deliveries.shape[0]
print("Number of rows:", num_rows)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_deliveries['batsman_runs'], bins=7, kde=False, color="blue")  
plt.title("Distribution of Runs per Ball")
plt.xlabel("Runs per Ball")
plt.ylabel("Frequency")
plt.xticks(range(7))
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_deliveries['batsman_runs'], bins=7, kde=True, color="blue")  
plt.title("Distribution of Runs per Ball")
plt.xlabel("Runs per Ball")
plt.ylabel("Frequency")
plt.xticks(range(7))
plt.show()
```

```python
team_wins = df["winner"].value_counts()
team_wins
```

```python
plt.figure(figsize=(12,6))
plt.bar(team_wins.index, team_wins.values, color='blue')
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Total Wins")
plt.title("Total Matches Won by Each Team in IPL")
plt.show()
```

```python
matches_per_season = df["season"].value_counts().sort_index()
matches_per_season
```

```python
plt.figure(figsize=(10,5))
plt.plot(matches_per_season.index, matches_per_season.values, marker='o', linestyle='-', color='green')
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Total Matches Played in Each IPL Season")
plt.xticks(matches_per_season.index, rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_deliveries["total_runs"], bins=10, color='gray', edgecolor='blue')
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.title("Distribution of Runs Scored Per Delivery")
plt.show()
```

```python
extras_type = df_deliveries["extras_type"].value_counts()
extras_type
```

```python
plt.figure(figsize=(6,6))
plt.pie(extras_type, labels=extras_type.index, autopct="%1.1f%%", colors=['gold', 'lightgreen'])
plt.title("Extras Type in IPL")
plt.show()
```

```python
win_by = df["result"].value_counts()
win_by
```

```python
plt.figure(figsize=(6,6))
plt.pie(win_by, labels=win_by.index, autopct="%1.1f%%", colors=['lightblue', 'lightgray'])
plt.title("win by runs/ wicket")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="batting_team", y="total_runs", data=df_deliveries, palette="pastel")

plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Runs Per Over")
plt.title("Runs Distribution Per Over for Each Team")
plt.show()
```

```python
toss_trends = df.groupby(["season", "toss_decision"]).size().unstack()
toss_trends.plot(kind="bar", stacked=True, figsize=(12,6), colormap="coolwarm")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Toss Decision Trends Over the Years")
plt.legend(title="Toss Decision")
plt.xticks(rotation=45)
plt.show()
```

```python
toss_vs_win = pd.crosstab(df["toss_decision"], df["winner"])
toss_vs_win
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(toss_vs_win, cmap="coolwarm", annot=False, fmt="d")
plt.xlabel("Winning Team")
plt.ylabel("Toss Decision")
plt.title("Impact of Toss Decision on Match Result")
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.heatmap(toss_vs_win, cmap="coolwarm", annot=True, fmt="d")
plt.xlabel("Winning Team")
plt.ylabel("Toss Decision")
plt.title("Impact of Toss Decision on Match Result")
plt.show()
```

```python
matches_per_season = df["season"].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.fill_between(matches_per_season.index, matches_per_season.values, color="skyblue", alpha=0.6)
plt.plot(matches_per_season.index, matches_per_season.values, marker="o", color="blue")
plt.xlabel("Season")
plt.ylabel("Matches Played")
plt.title("Total Matches Played Per Season")
plt.grid()
plt.show()
```

```python
toss_decisions = df["toss_decision"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(toss_decisions, labels=toss_decisions.index, autopct="%1.1f%%", colors=['gold', 'lightblue'], wedgeprops=dict(width=0.3))
plt.title("Toss Decision Ratio in IPL")
plt.show()
```

```python
top_batsmen = df_deliveries.groupby("batter")["batsman_runs"].sum().nlargest(10)

# Plot
plt.figure(figsize=(10,5))
plt.bar(top_batsmen.index, top_batsmen.values, color='blue')
plt.xlabel("Batsmen")
plt.ylabel("Total Runs")
plt.title("Top 10 Run Scorers in IPL")
plt.xticks(rotation=45)
plt.show()
```
